```python
import pandas as pd
from pandas import read_csv
import matplotlib.pyplot as plt
```

```python
path = '/Users/mariabelenalberti/Documents/smart-meters-in-london'
filename = '/weather_hh.csv'

weather = read_csv(path+filename)
```

```python
# Extract the month and year for ease of analysis
weather['timestmap']= pd.to_datetime(weather['timestamp'],format = '%d/%m/%Y %H:%M')
weather['year'] = pd.DatetimeIndex(weather['timestmap']).year
weather['month'] = pd.DatetimeIndex(weather['timestmap']).month
weather['Date'] = pd.to_datetime(weather.year.astype(str) + '-' + weather.month.astype(str))
```

```python
del weather['year']
del weather['month']
del weather['timestamp']
```

```python
weather_group_by = weather.groupby(['Date']).mean().reset_index()
```

```python
weather_group_by
```

```python
path = '/Users/mariabelenalberti/Documents/smart-meters-in-london/daily_dataset.csv/'
filename = 'daily_dataset.csv'

energy = read_csv(path+filename)
```

```python
energy.head()
```

```python
del energy['energy_median']
del energy['energy_max']
del energy['energy_count']
del energy['energy_std']
del energy['energy_sum']
del energy['energy_min']
```

```python
energy.head()
```

```python
energy['day']= pd.to_datetime(energy['day'],format = '%Y-%m-%d')
energy['year'] = pd.DatetimeIndex(energy['day']).year
energy['month'] = pd.DatetimeIndex(energy['day']).month
energy['Date'] = pd.to_datetime(energy.year.astype(str) + '-' + energy.month.astype(str))
```

```python
energy.head()
```

```python
energy_12_13 = energy[(energy['year'] == 2012) | (energy['year'] == 2013)]
```

```python
energy_12_13.head()
```

```python
del energy_12_13['day']
del energy_12_13['month']
del energy_12_13['LCLid']
del energy_12_13['year']
```

```python
energy_12_13.head()
```

```python
energy_group = energy_12_13.groupby(['Date']).mean().reset_index()
```

```python
energy_weather = pd.merge(weather_group_by,energy_group,on = 'Date',how = 'left')
```

```python
energy_weather
```
