```python
import pandas as pd
from pandas import read_csv
import matplotlib.pyplot as plt
import glob
```

```python
path = '/Users/mariabelenalberti/OneDrive - Universidad Torcuato Di Tella/Tesis/code_datasets/1_data_frames/univariate/affluent_std/'
filename = 'MAC004463.csv'
MAC004463 = read_csv(path+filename)
```

```python
MAC004463['tstp']= pd.to_datetime(MAC004463['tstp'],format = '%d/%m/%Y %H:%M')
MAC004463['year'] = pd.DatetimeIndex(MAC004463['tstp']).year
```

```python
MAC004463['month'] = pd.DatetimeIndex(MAC004463['tstp']).month
MAC004463['Date'] = pd.to_datetime(MAC004463.year.astype(str) + '-' + MAC004463.month.astype(str))
```

```python
MAC004463
```

```python
MAC004463_jan_12 = MAC004463[MAC004463['Date'] == '2012-01-01']
MAC004463_feb_12 = MAC004463[MAC004463['Date'] == '2012-02-01']
MAC004463_mar_12 = MAC004463[MAC004463['Date'] == '2012-03-01']
```

```python
MAC004463_jan_12['day'] = pd.DatetimeIndex(MAC004463_jan_12['tstp']).day
```
